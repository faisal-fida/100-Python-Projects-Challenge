```python
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import fetch_california_housing, load_iris, load_diabetes
boston = fetch_california_housing()
diabetes = load_diabetes()
iris = load_iris()
```

Line graphs are used to show how a variable changes over time. They are often used to visualize the training progress of a machine learning model.

```python
plt.plot(boston.data[:, 0], boston.data[:, 5])
plt.xlabel('Year')
plt.ylabel('Median home value ($)')
plt.show()
```

Bar graphs are used to compare the values of different variables. They are often used to visualize the performance of different machine learning models on a given task.

```python
models = ['Linear regression', 'Decision tree', 'Random forest']
accuracies = [0.85, 0.90, 0.95]

sns.barplot(x=models, y=accuracies)
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.show()
```

Histograms are used to show the distribution of a single variable. They are often used to visualize the distribution of the features in a machine learning dataset.

```python
plt.hist(iris.data[:, 0])
plt.xlabel('Sepal width (cm)')
plt.ylabel('Count')
plt.show()
```

Scatter plots are used to show the relationship between two variables. They are often used to visualize the relationship between the features and the target variable in a machine learning dataset.

```python
plt.scatter(diabetes.data[:, 0], diabetes.data[:, 1])
plt.xlabel('Blood sugar levels')
plt.ylabel('Insulin levels')
plt.show()
```

Heatmaps are used to show the correlation between multiple variables. They are often used to visualize the correlation between the features in a machine learning dataset.

```python
plt.figure(figsize=(100, 100))
sns.heatmap(iris.data, annot=True)
plt.show()
```
